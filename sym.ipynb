```python
import poppy
import jax
import functools
```

```python
n = 4
N = poppy.factorial(n)

print(f'\nTHE SYMMETRIC GROUP S{n}\n\n{poppy.sym(n,N)}')
```

```python
n = 5
N = poppy.factorial(n)
TABLE = poppy.symtbl(n,N)
TITLE = f'S{n} MULTIPLICATION TABLE'

poppy.plot(TABLE, title = TITLE, size = 12)
```

```python
f = poppy.field(3,2)

for p in prts[8:9]:
  prt = jax.numpy.array(p, dtype = jax.numpy.int8)
  rep = poppy.symrep(prt,f)
  print(f'\nTHE {prt} REPRESENTATION OF S{prt.sum().item()}\n\n{rep.proj().squeeze()}\n\n')
```
